# 3 概率与统计

## 3.1 理论计算题 - 贝叶斯公式

某疾病在人群中的患病率为 0.1%。现有一种检测方法：
- 若患病，检测呈阳性的概率为 99%（灵敏度）
- 若未患病，检测呈阳性的概率为 2%（假阳性率）
一个人检测结果为阳性，求他真正患病的概率（使用贝叶斯公式）。

```python
# 贝叶斯公式计算
# P(患病|阳性) = P(阳性|患病) * P(患病) / P(阳性)

# 已知概率
P_disease = 0.001      # 患病率 0.1%
P_no_disease = 1 - P_disease  # 未患病率 99.9%
P_positive_given_disease = 0.99  # 灵敏度 99%
P_positive_given_no_disease = 0.02  # 假阳性率 2%

# P(阳性) = P(阳性|患病) * P(患病) + P(阳性|未患病) * P(未患病)
P_positive = (P_positive_given_disease * P_disease + 
              P_positive_given_no_disease * P_no_disease)

# 贝叶斯公式: P(患病|阳性) = P(阳性|患病) * P(患病) / P(阳性)
P_disease_given_positive = (P_positive_given_disease * P_disease) / P_positive

print('贝叶斯公式计算结果:')
print(f'P(阳性) = {P_positive:.6f}')
print(f'P(患病|阳性) = {P_disease_given_positive:.6f}')
print(f'即: 检测结果为阳性时，真正患病的概率约为 {P_disease_given_positive*100:.4f}%')
```

## 3.2 编程题 - 中心极限定理模拟

```python
%matplotlib inline
import matplotlib.pyplot as plt

# 1. 生成 n = 10000 个服从均匀分布 U(0, 1) 的随机变量，重复 m = 1000 次
# 每次计算这 n 个变量的均值

n = 10000  # 每次采样的样本数
m = 1000   # 重复次数

# 存储每次的均值
means = []

for i in range(m):
    sample = np.random.uniform(0, 1, n)  # 生成 n 个均匀分布样本
    sample_mean = np.mean(sample)         # 计算均值
    means.append(sample_mean)

means = np.array(means)
print('1. 中心极限定理模拟完成')
print(f'   生成了 {m} 个均值，每个均值基于 {n} 个均匀分布样本')
print(f'   均值的均值 = {np.mean(means):.6f}')
print(f'   均值的期望 ≈ 0.5 (理论值)')
```

```python
# 2. 绘制这 1000 个均值的直方图，并叠加理论正态分布的概率密度曲线

# 计算理论均值和方差
theoretical_mean = 0.5
theoretical_var = (1/12) / n  # 均匀分布 U(0,1) 的方差为 1/12，除以 n
theoretical_std = np.sqrt(theoretical_var)

# 绘制直方图
plt.figure(figsize=(10, 6))
plt.hist(means, bins=30, density=True, alpha=0.7, color='blue', edgecolor='black', label='均值的直方图')

# 叠加理论正态分布曲线
x = np.linspace(means.min(), means.max(), 100)
normal_pdf = (1 / (theoretical_std * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - theoretical_mean) / theoretical_std) ** 2)
plt.plot(x, normal_pdf, 'r-', linewidth=2, label=f'理论正态分布 N({theoretical_mean}, {theoretical_var:.6f})')

plt.xlabel('均值')
plt.ylabel('概率密度')
plt.title('中心极限定理验证 - 1000个样本均值的分布')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print('2. 直方图和理论正态分布曲线已绘制')
```

```python
# 3. 计算这些均值的实际方差
actual_variance = np.var(means)
theoretical_variance = (1/12) / n

print('3. 方差计算结果:')
print(f'   实际方差 = {actual_variance:.8f}')
print(f'   理论方差 = {theoretical_variance:.8f}')
print(f'   差异 = {abs(actual_variance - theoretical_variance):.8f}')
```

# 4 导数、反向传播和复杂度

## 4.1 理论计算题

给定函数 z = (w₁x₁ + w₂x₂ - y)²，其中 x₁ = 2, x₂ = 1, y = 3 为常数，w₁, w₂ 为变量。
1. 分别计算 ∂z/∂w₁ 和 ∂z/∂w₂ 的表达式。
2. 若当前 w₁ = 0.5, w₂ = 1，计算梯度值。

```python
# 4.1 理论计算题 - 导数计算

# 给定参数
x1 = 2
x2 = 1
y = 3
w1 = 0.5
w2 = 1

# 1. 计算偏导数表达式
# z = (w1*x1 + w2*x2 - y)^2
# ∂z/∂w1 = 2*(w1*x1 + w2*x2 - y)*x1
# ∂z/∂w2 = 2*(w1*x1 + w2*x2 - y)*x2

# 2. 计算梯度值
def compute_gradient(w1, w2, x1, x2, y):
    # 计算中间值
    linear = w1 * x1 + w2 * x2 - y
    # 计算偏导数
    dz_dw1 = 2 * linear * x1
    dz_dw2 = 2 * linear * x2
    return dz_dw1, dz_dw2

# 计算梯度
dz_dw1, dz_dw2 = compute_gradient(w1, w2, x1, x2, y)

print('4.1 理论计算题结果:')
print(f'1. 偏导数表达式:')
print(f'   ∂z/∂w1 = 2*(w1*2 + w2*1 - 3)*2 = 4*(2w1 + w2 - 3)')
print(f'   ∂z/∂w2 = 2*(w1*2 + w2*1 - 3)*1 = 2*(2w1 + w2 - 3)')
print()
print(f'2. 当 w1 = {w1}, w2 = {w2} 时的梯度值:')
print(f'   ∂z/∂w1 = {dz_dw1}')
print(f'   ∂z/∂w2 = {dz_dw2}')
```

## 4.2 编程题 - 手动实现反向传播

手动实现一个简单计算图的反向传播:
1. 定义前向计算：输入 x, 经过 a = x·w₁, b = a + w₂, 输出 L = b²。
2. 给定 x = 2, w₁ = 1.5, w₂ = 0.5，计算前向结果。
3. 手动（按链式法则）计算损失 L 对 w₁, w₂ 的梯度，并验证与自动微分工具（如 torch.autograd）的结果是否一致。

```python
# 4.2 编程题 - 手动实现反向传播

# 给定参数
x = 2
w1 = 1.5
w2 = 0.5

# 1. 前向计算
a = x * w1
b = a + w2
L = b ** 2

# 2. 手动计算梯度（链式法则）
# ∂L/∂b = 2b
# ∂L/∂a = ∂L/∂b * ∂b/∂a = 2b * 1 = 2b
# ∂L/∂w1 = ∂L/∂a * ∂a/∂w1 = 2b * x
# ∂L/∂w2 = ∂L/∂b * ∂b/∂w2 = 2b * 1 = 2b

dL_db = 2 * b
dL_dw1 = dL_db * x
dL_dw2 = dL_db * 1

print('4.2 编程题 - 手动实现反向传播:')
print(f'1. 前向计算结果:')
print(f'   a = x * w1 = {x} * {w1} = {a}')
print(f'   b = a + w2 = {a} + {w2} = {b}')
print(f'   L = b² = {b}² = {L}')
print()
print(f'2. 手动计算梯度 (链式法则):')
print(f'   ∂L/∂w1 = 2b * x = 2*{b}*{x} = {dL_dw1}')
print(f'   ∂L/∂w2 = 2b = 2*{b} = {dL_dw2}')
```

```python
# 3. 使用 PyTorch 自动微分验证
try:
    import torch
    
    # 定义变量并设置 requires_grad=True
    x_tensor = torch.tensor(x, dtype=torch.float32)
    w1_tensor = torch.tensor(w1, dtype=torch.float32, requires_grad=True)
    w2_tensor = torch.tensor(w2, dtype=torch.float32, requires_grad=True)
    
    # 前向计算
    a_tensor = x_tensor * w1_tensor
    b_tensor = a_tensor + w2_tensor
    L_tensor = b_tensor ** 2
    
    # 反向传播
    L_tensor.backward()
    
    # 获取梯度
    auto_dL_dw1 = w1_tensor.grad.item()
    auto_dL_dw2 = w2_tensor.grad.item()
    
    print()
    print('3. PyTorch 自动微分验证:')
    print(f'   自动微分 ∂L/∂w1 = {auto_dL_dw1}')
    print(f'   自动微分 ∂L/∂w2 = {auto_dL_dw2}')
    print()
    print('验证结果:')
    print(f'   ∂L/∂w1 手动计算与自动微分是否一致: {abs(dL_dw1 - auto_dL_dw1) < 1e-6}')
    print(f'   ∂L/∂w2 手动计算与自动微分是否一致: {abs(dL_dw2 - auto_dL_dw2) < 1e-6}')
except ImportError:
    print()
    print('3. PyTorch 未安装，无法进行自动微分验证。请运行: pip install torch')
    print('   手动计算的梯度结果仍然有效。')
```

# 5 线性方法、基础优化和 softmax 回归

## 5.1 理论计算题

考虑线性回归模型 y = wx + b，使用均方误差损失
L = (1/n) * Σ(y_i - (w x_i + b))²
推导损失对 w 和 b 的偏导数表达式（用求和形式）。

```python
# 5.1 理论计算题 - 线性回归偏导数推导

print('5.1 理论计算题 - 线性回归偏导数推导:')
print('')
print('给定均方误差损失函数:')
print('L = (1/n) * Σ(y_i - (w x_i + b))²')
print('')
print('1. 对 w 的偏导数:')
print('dL/dw = (1/n) * Σ 2*(y_i - (w x_i + b))*(-x_i)')
print('      = -(2/n) * Σ x_i*(y_i - (w x_i + b))')
print('')
print('2. 对 b 的偏导数:')
print('dL/db = (1/n) * Σ 2*(y_i - (w x_i + b))*(-1)')
print('      = -(2/n) * Σ (y_i - (w x_i + b))')
print('')
print('其中:')
print('- n 是样本数量')
print('- Σ 表示从 i=1 到 n 的求和')
print('- y_i 是第 i 个样本的真实值')
print('- x_i 是第 i 个样本的输入特征')
print('- w 是权重参数')
print('- b 是偏置参数')
```

## 5.2 编程题 - 从零实现 softmax 回归

从零实现 softmax 回归对手写数字 (MNIST) 进行分类:
1. 使用 sklearn.datasets.load_digits() 或自己加载 MNIST 子集。
2. 将图像展平为向量，对标签进行 one-hot 编码。
3. 实现 softmax 函数、交叉熵损失函数。
4. 使用小批量随机梯度下降（批量大小 32，学习率 0.1）训练模型 50 个 epoch。
5. 输出测试集上的准确率。

```python
# 5.2 编程题 - 从零实现 softmax 回归
import numpy as np
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split

# 1. 加载数据集
print('1. 加载数据集...')
digits = load_digits()
X = digits.data  # 特征 (n_samples, n_features)
y = digits.target  # 标签 (n_samples,)

print(f'数据集大小: {X.shape[0]} 样本, {X.shape[1]} 特征')
print(f'标签范围: {np.min(y)} - {np.max(y)}')

# 2. 数据预处理
print('\n2. 数据预处理...')

# 归一化特征
X = X / 16.0  # 将像素值归一化到 [0, 1]

# 对标签进行 one-hot 编码
def one_hot_encode(y, num_classes):
    n = y.shape[0]
    one_hot = np.zeros((n, num_classes))
    one_hot[np.arange(n), y] = 1
    return one_hot

num_classes = len(np.unique(y))
y_one_hot = one_hot_encode(y, num_classes)

# 划分训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(X, y_one_hot, test_size=0.2, random_state=42)
y_train_labels = np.argmax(y_train, axis=1)
y_test_labels = np.argmax(y_test, axis=1)

print(f'训练集: {X_train.shape[0]} 样本')
print(f'测试集: {X_test.shape[0]} 样本')
```

```python
# 3. 实现 softmax 函数和交叉熵损失函数
print('\n3. 实现核心函数...')

def softmax(z):
    # 数值稳定性处理：减去最大值
    exp_z = np.exp(z - np.max(z, axis=1, keepdims=True))
    return exp_z / np.sum(exp_z, axis=1, keepdims=True)

def cross_entropy_loss(y_pred, y_true):
    # y_pred: (batch_size, num_classes)
    # y_true: (batch_size, num_classes)
    n = y_pred.shape[0]
    return -np.mean(np.sum(y_true * np.log(y_pred + 1e-10), axis=1))

# 4. 初始化模型参数
n_features = X_train.shape[1]
n_classes = y_train.shape[1]

# 随机初始化权重和偏置
np.random.seed(42)
W = np.random.randn(n_features, n_classes) * 0.01
b = np.zeros(n_classes)

print(f'模型参数: W({n_features}x{n_classes}), b({n_classes})')
```

```python
# 5. 实现小批量随机梯度下降
print('\n4. 训练模型...')

# 超参数
batch_size = 32
learning_rate = 0.1
epochs = 50

def compute_gradient(X_batch, y_batch, W, b):
    # 前向传播
    z = X_batch @ W + b
    y_pred = softmax(z)
    
    # 计算损失
    loss = cross_entropy_loss(y_pred, y_batch)
    
    # 反向传播计算梯度
    n = X_batch.shape[0]
    grad_z = y_pred - y_batch  # (batch_size, num_classes)
    grad_W = (X_batch.T @ grad_z) / n
    grad_b = np.mean(grad_z, axis=0)
    
    return loss, grad_W, grad_b

# 训练循环
train_losses = []

for epoch in range(epochs):
    # 打乱训练数据
    indices = np.random.permutation(X_train.shape[0])
    X_train_shuffled = X_train[indices]
    y_train_shuffled = y_train[indices]
    
    batch_losses = []
    
    # 小批量训练
    for i in range(0, X_train.shape[0], batch_size):
        # 获取当前批次
        end = min(i + batch_size, X_train.shape[0])
        X_batch = X_train_shuffled[i:end]
        y_batch = y_train_shuffled[i:end]
        
        # 计算梯度
        loss, grad_W, grad_b = compute_gradient(X_batch, y_batch, W, b)
        batch_losses.append(loss)
        
        # 更新参数
        W -= learning_rate * grad_W
        b -= learning_rate * grad_b
    
    # 计算 epoch 平均损失
    epoch_loss = np.mean(batch_losses)
    train_losses.append(epoch_loss)
    
    # 每 10 个 epoch 打印一次
    if (epoch + 1) % 10 == 0:
        print(f'Epoch {epoch+1}/{epochs}, Loss: {epoch_loss:.4f}')

print('训练完成!')
```

```python
# 6. 评估模型
print('\n5. 评估模型...')

def predict(X, W, b):
    z = X @ W + b
    y_pred = softmax(z)
    return np.argmax(y_pred, axis=1)

# 在测试集上预测
y_pred_test = predict(X_test, W, b)

# 计算准确率
accuracy = np.mean(y_pred_test == y_test_labels)
print(f'测试集准确率: {accuracy:.4f}')

# 打印前 10 个预测结果
print('\n前 10 个测试样本的预测结果:')
for i in range(10):
    print(f'Sample {i+1}: 真实标签 = {y_test_labels[i]}, 预测标签 = {y_pred_test[i]}')
```

# 6 最大似然估计和逻辑回归

## 6.1 理论计算题

假设样本 x₁, x₂, ..., xₙ 独立同分布于正态分布 N(μ, σ²)。
1. 写出似然函数 L(μ, σ²)。
2. 证明：μ 的最大似然估计为 μ̂ = (1/n)Σxᵢ。
3. 证明：σ² 的最大似然估计为 σ̂² = (1/n)Σ(xᵢ - μ̂)²。

```python
# 6.1 理论计算题 - 最大似然估计推导

print('6.1 理论计算题 - 正态分布的最大似然估计:')
print('')
print('1. 似然函数 L(μ, σ²):')
print('L(μ, σ²) = ∏ (1/(√(2πσ²))) * exp(-(x_i - μ)²/(2σ²)) for i=1 to n')
print('        = (2πσ²)^(-n/2) * exp(-(1/(2σ²))Σ(x_i - μ)²)')
print('')
print('2. 对 μ 求最大似然估计:')
print('取对数似然:')
print('log L(μ, σ²) = -n/2 * log(2πσ²) - (1/(2σ²))Σ(x_i - μ)²')
print('')
print('对 μ 求偏导并令其为 0:')
print('∂log L/∂μ = (1/σ²)Σ(x_i - μ) = 0')
print('解得: Σx_i - nμ = 0 → μ̂ = (1/n)Σx_i')
print('')
print('3. 对 σ² 求最大似然估计:')
print('将 μ = μ̂ 代入对数似然，对 σ² 求偏导并令其为 0:')
print('∂log L/∂σ² = -n/(2σ²) + (1/(2σ^4))Σ(x_i - μ̂)² = 0')
print('解得: σ̂² = (1/n)Σ(x_i - μ̂)²')
print('')
print('结论:')
print('- μ 的最大似然估计为样本均值')
print('- σ² 的最大似然估计为样本方差（除以 n，无偏估计除以 n-1）')
```

## 6.2 编程题 - 实现二分类逻辑回归模型

实现二分类逻辑回归模型（不使用深度学习框架的高阶 API）:
1. 生成两类线性可分的数据：每类 200 个样本，特征维度为 2。
2. 定义 sigmoid 函数和二元交叉熵损失。
3. 使用梯度下降法（学习率 0.1, 迭代 1000 次）学习参数 w,b。
4. 绘制决策边界，并计算模型在测试集（新生成 100 个样本）上的分类准确率。

```python
# 6.2 编程题 - 实现二分类逻辑回归模型
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

# 1. 生成两类线性可分的数据
print('1. 生成数据...')

np.random.seed(42)
n_samples = 200  # 每类样本数

# 第一类数据：均值 (2, 3)，方差 1
class1 = np.random.randn(n_samples, 2) + np.array([2, 3])
y1 = np.zeros(n_samples)  # 标签 0

# 第二类数据：均值 (5, 1)，方差 1
class2 = np.random.randn(n_samples, 2) + np.array([5, 1])
y2 = np.ones(n_samples)   # 标签 1

# 合并数据
X = np.vstack([class1, class2])
y = np.hstack([y1, y2])

# 打乱数据
indices = np.random.permutation(len(X))
X = X[indices]
y = y[indices]

print(f'数据形状: X={X.shape}, y={y.shape}')
print(f'第一类样本数: {sum(y==0)}')
print(f'第二类样本数: {sum(y==1)}')

```

```python
# 2. 定义 sigmoid 函数和二元交叉熵损失
print('\n2. 定义核心函数...')

def sigmoid(z):
    # 避免数值溢出
    return 1 / (1 + np.exp(-np.clip(z, -250, 250)))

def binary_cross_entropy(y_pred, y_true):
    # y_pred: (n_samples,), y_true: (n_samples,)
    epsilon = 1e-10
    return -np.mean(y_true * np.log(y_pred + epsilon) + (1 - y_true) * np.log(1 - y_pred + epsilon))

# 3. 初始化参数
n_features = X.shape[1]
w = np.random.randn(n_features) * 0.01
b = 0.0

print(f'初始参数: w={w}, b={b}')
```

```python
# 4. 使用梯度下降法学习参数
print('\n3. 训练模型...')

# 超参数
learning_rate = 0.1
iterations = 1000

def compute_gradient(X, y, w, b):
    # 前向传播
    z = X @ w + b
    y_pred = sigmoid(z)
    
    # 计算损失
    loss = binary_cross_entropy(y_pred, y)
    
    # 计算梯度
    n = len(X)
    grad_w = (1/n) * X.T @ (y_pred - y)
    grad_b = (1/n) * np.sum(y_pred - y)
    
    return loss, grad_w, grad_b

# 训练过程
losses = []

for i in range(iterations):
    # 计算梯度
    loss, grad_w, grad_b = compute_gradient(X, y, w, b)
    losses.append(loss)
    
    # 更新参数
    w -= learning_rate * grad_w
    b -= learning_rate * grad_b
    
    # 每 100 次迭代打印一次
    if (i + 1) % 100 == 0:
        print(f'Iteration {i+1}/{iterations}, Loss: {loss:.4f}')

print('训练完成!')
print(f'最终参数: w={w}, b={b}')

```

```python
# 5. 绘制决策边界
print('\n4. 绘制决策边界...')

# 生成网格点
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                     np.linspace(y_min, y_max, 100))

# 预测网格点
Z = sigmoid(np.c_[xx.ravel(), yy.ravel()] @ w + b)
Z = Z.reshape(xx.shape)

# 绘制决策边界
plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap='RdBu')
plt.contour(xx, yy, Z, levels=[0.5], colors='black', linewidths=2)
plt.scatter(X[y==0, 0], X[y==0, 1], c='blue', label='Class 0', alpha=0.6)
plt.scatter(X[y==1, 0], X[y==1, 1], c='red', label='Class 1', alpha=0.6)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Decision Boundary')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# 6. 生成测试数据并评估模型
print('\n5. 评估模型...')

# 生成测试数据
n_test = 100
test_class1 = np.random.randn(n_test//2, 2) + np.array([2, 3])
test_y1 = np.zeros(n_test//2)
test_class2 = np.random.randn(n_test//2, 2) + np.array([5, 1])
test_y2 = np.ones(n_test//2)

X_test = np.vstack([test_class1, test_class2])
y_test = np.hstack([test_y1, test_y2])

# 预测
def predict(X, w, b, threshold=0.5):
    z = X @ w + b
    y_pred = sigmoid(z)
    return (y_pred >= threshold).astype(int)

y_pred_test = predict(X_test, w, b)

# 计算准确率
accuracy = np.mean(y_pred_test == y_test)
print(f'测试集准确率: {accuracy:.4f}')

# 打印前 10 个预测结果
print('\n前 10 个测试样本的预测结果:')
for i in range(10):
    print(f'Sample {i+1}: 真实标签 = {int(y_test[i])}, 预测标签 = {y_pred_test[i]}')
```
